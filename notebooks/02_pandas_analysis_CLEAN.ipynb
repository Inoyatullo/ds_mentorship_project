```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
print('✅ Библиотеки загружены')
```

```python
# Загрузка данных Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(f'✅ Данные загружены: {df.shape[0]} строк, {df.shape[1]} столбцов')
print('Первые 5 строк:')
df.head()
```

### Анализ пропусков
1. Больше всего пропусков в колонках: `Cabin` (77.10%) и `Age` (19.87%)

### Анализ выживаемости  
1. Самый высокий шанс выжить: **женщины 1 класса** (96.81%)
2. Самый низкий шанс выжить: **мужчины 3 класса** (13.54%)

### Анализ графиков
1-й и 3-й графики наиболее наглядно показывают разницу между группами.

```python
# Анализ пропусков
missing_info = pd.DataFrame({
    'Пропуски': df.isnull().sum(),
    'Доля, %': (df.isnull().sum() / len(df) * 100).round(2)
})
missing_data = missing_info[missing_info['Пропуски'] > 0]
missing_data
```

```python
# Создаем копию датафрейма для экспериментов
df_engineered = df.copy()

# 1. Общий размер семьи
df_engineered['FamilySize'] = df_engineered['SibSp'] + df_engineered['Parch'] + 1

# 2. Одинокий пассажир
df_engineered['IsAlone'] = (df_engineered['FamilySize'] == 1).astype(int)

# Посмотрим на выживаемость в новых группах
print('Выживаемость в зависимости от размера семьи:')
print(df_engineered.groupby('FamilySize')['Survived'].mean().round(3))

print('\nВыживаемость одиноких пассажиров:')
print(df_engineered.groupby('IsAlone')['Survived'].mean().round(3))
```

```python
# Сохраняем датасет с новыми признаками
df_engineered.to_csv('data/titanic_processed.csv', index=False)
print('✅ Обработанные данные сохранены в data/titanic_processed.csv')
```
